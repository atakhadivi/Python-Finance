```python
import matplotlib.pyplot as plt
```

```python
# Dictionary to store budget 
budget = {}
```

```python
# Function to add income  
def add_income(amount, source):
    budget[source] = amount  
```

```python
# Function  to add new expense
def add_expense(amount, category):    
    if category not in budget:
        budget[category] = -amount
    else:
        budget[category] -= amount
```

```python

# Get income amounts  
num_incomes = int(input("How many income sources do you have? "))  

for i in range(num_incomes):
    source = input("Enter source name: ")
    amount = float(input("Enter amount: "))
    add_income(amount, source)

# Get expenses
num_expenses = int(input("How many expenses do you have? "))  

for i in range(num_expenses):
    category = input("Enter category: ")
    amount = float(input("Enter amount: ")) 
    add_expense(amount, category)
```

```python
# Separate incomes and expenses
incomes = {k:v for k,v in budget.items() if v > 0}
expenses = {k:v for k,v in budget.items() if v < 0}

# Plot incomes and expenses in stacked bar chart
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(6, 8))

ax1.bar(incomes.keys(), incomes.values())
ax2.bar(expenses.keys(), expenses.values())

ax1.set_title('Incomes')
ax2.set_title('Expenses')

# Calculate and display total savings
total_income = sum(incomes.values())
total_expenses = sum(expenses.values()) * -1
savings = total_income - total_expenses

print(f"Total Income: {total_income:,}")
print(f"Total Expenses: {total_expenses:,}")
print(f"Total Savings: {savings:,}")
```
